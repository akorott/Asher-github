## Scale and reshape

```python
# The current range of X_train and X_test is 0 to 255.
# The code below is equivalent to X_train = X_train / 255.
# This scales each value to be between 0 and 1.

x_train/=255
x_test/=255
```

```python
x_train.iloc[1]
```

```python
x_train = x_train.values.reshape(-1, 28, 28, 1)
x_test = x_test.values.reshape(-1, 28, 28, 1)
```

```python
print(x_train.shape, x_test.shape)
```

```python
y_train = np_utils.to_categorical(y_train)
```

```python
y_train[:5]
```

```python
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
```

```python
x_train[0]
```

## CNN with randomized search

```python
def model_func(filter_one=6, kernel_size_one=3, poolsize_one=(2,2), filter_two=8, kernel_size_two=3, neurons=128):
    # Instantiate a CNN.
    cnn = Sequential()
    #convolution layer
    cnn.add(Conv2D(filters = filter_one, kernel_size = kernel_size_one, activation='relu', input_shape=(28, 28, 1)))
    # dimensions of region of pooling
    cnn.add(MaxPooling2D(pool_size = poolsize_one))
    # Add a convolutional layer with filters of 
    # size 3x3 and the ReLU activation function. 
    cnn.add(Conv2D(filters=filter_two, kernel_size = kernel_size_two,activation='relu'))
    # Flatten our result so that we can pass it into a dense layer.
    cnn.add(Flatten())
    # Add a densely-connected layer with x number neurons.
    cnn.add(Dense(neurons, activation='relu'))
    # Add a final layer with 10 neurons.
    cnn.add(Dense(10, activation='softmax'))
    # Compile model
    cnn.compile(loss='categorical_crossentropy',
                      optimizer='adam',
                      metrics=['accuracy'])
    return cnn
```

```python
nn = KerasClassifier(build_fn=model_func, epochs=5, batch_size=256, verbose=0)
```

```python
params = {
    'filter_one':[10, 12, 16],
    'kernel_size_one':[5,7],
    'poolsize_one':[(2,2)],
    'filter_two':[64, 32],
    'kernel_size_two':[3],
    'neurons':[256, 512]
}
```

```python
rs = RandomizedSearchCV(nn, param_distributions = params)
rs.fit(x_train,y_train)
print(rs.best_score_)
print(rs.best_params_)
```

```python
rs.score(x_train,y_train)
```

## Prediction time!
